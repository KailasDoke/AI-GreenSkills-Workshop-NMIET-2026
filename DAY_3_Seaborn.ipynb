```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "Solar": [1200,1500,1300],
    "Wind": [3400,3600,3200],
    "Hydropower":[2900,3100,2800],
    "Biomass" :[2500,2700,2400]
}
df = pd.DataFrame(data)
sns.pairplot(df)
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.9)
plt.title("Correlation Between Energy Source")
plt.show()
```

```python
sns.boxplot(data=df)
#box plot
plt.title("Distributed od Energy Consumption by Source")
plt.show()
```

```python
# create a violin plot
sns.violinplot(data=df)
plt.title("Distributed od Energy Consumption by Source")
plt.show()
```

```python
#
energy_values = [ 100,200,300,400,500]
carbon_emissions = [10,20,30,40,50]

df_reg = pd.DataFrame({
    "Energy Consumption ": energy_values,
    "Carbon Emissions": carbon_emissions
})

sns.regplot(x="Energy Consumption ", y="Carbon Emissions", data=df_reg)
plt.title("Energy Consumption vs Carbon Emissions")
plt.xlabel("Energy Consumption")
plt.ylabel("Carbon Emissions")
plt.show()
```
