```python
total_consumption = np.sum(energy_consumption)
print(f"total Energy Consumption: {total_consumption}mwh")
```

```python
total_consumption = np.mean(energy_consumption)#mean
print(f"total Energy Consumption: {total_consumption:.2f}mwh")
```

```python
total_consumption = np.std(energy_consumption)
print(f"total Energy Consumption: {total_consumption:.2f}mwh")
```
