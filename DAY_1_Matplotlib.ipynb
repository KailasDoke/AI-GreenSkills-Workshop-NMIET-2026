```python
import matplotlib . pyplot as plt
months = ["jan","feb","mar","apr","may","jun"]
energy_consumption = [2500,2600,2700,1100,200,3545]
plt.plot(months,energy_consumption, marker='o',color='k',linestyle='--')
plt.title("Energy Consumption over 6 Months")
plt.xlabel("Month")
plt.ylabel("Energy consumption")
plt.show()
```

```python
import matplotlib . pyplot as plt
months = ["jan","feb","mar","apr","may","jun"]#BAR CHART
energy_consumption = [2500,2600,2700,1100,200,3545]
plt.bar(months,energy_consumption,color='k')
plt.title("Energy Consumption over 6 Months")
plt.xlabel("Month")
plt.ylabel("Energy consumption")
plt.show()
```

```python
plt.pie(energy_consumption,labels=energy_consumption,autopct='%1.1f%%')

plt.title("Energy Consumption by source")
plt.show()
```
