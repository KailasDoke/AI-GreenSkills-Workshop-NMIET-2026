```python
#dataset
climate_data = [
    {"city":"City A","temperature":25 ,"carbon_footprint":500},
    {"city":"City B","temperature":30 ,"carbon_footprint":350},
    {"city":"City C","temperature":40 ,"carbon_footprint":400},
    {"city":"City D","temperature":45 ,"carbon_footprint":200},
    {"city":"City E","temperature":45 ,"carbon_footprint":320},
]
high_tem_threshold = 30

high_temp_cities = [city for city in climate_data if city["temperature"]> high_tem_threshold]
print("hight temperature (>30)")
for city in high_temp_cities:
  print(f"{city['city']} : {city['temperature']}°C")
```
